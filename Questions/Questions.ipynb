```python
# Decorators
    # Decorator is a design and pattern in python , that is used to add new functionalities into the existing function without modifiying its code
    # this means when you call a decorator function that take another function as an arguments
    # it returns a new function that modifies the behaviour of the existing function

def my_decorator(func):
    def wrapper():
        print('test1')
        func()
        print('test2')
    return wrapper

@my_decorator
def test():
    print('test3')

test()
```

```python
# Polymorphism
    # also, self and other both take objects, whereas function take object and values


class Point:
    def __init__(self, a, b):
        self.a = a
        self.b = b
   
    def __add__(self, other):
        return Point(self.a + other.a , self.b + other.b)

test = Point(3,4)
test1 = Point(8,9)

run = (test + test1)  # it actually does  (test).__add__ (test1)

print(run.a)


# print(run)


```
