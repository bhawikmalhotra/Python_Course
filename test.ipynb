```python
# 7: From a mixed list (numbers + strings), create two separate lists â€” one for numbers and one for strings.

list1= [1,2,"age",5,6,'name', 'delhi', 75]

str_list =[]
int_list = []

for i in list1:
    if type(i) == str:
        str_list.append(i)
    else:
        int_list.append(i)

print(str_list)
print(int_list)
```
